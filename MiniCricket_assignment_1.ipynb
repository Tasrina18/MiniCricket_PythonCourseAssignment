```python
import random

# Toss
my_option = input("Enter your preference for toss (H/T): ").upper()

toss = random.choice(['H', 'T'])
print("Toss result:", toss)

my_run_array = []
computer_run_array = []


# FUNCTION: Batting Logic

def bat_first(batter_array, bowler_array, user_is_batter):
    """
    If user_is_batter = True → you are batting
    If user_is_batter = False → computer is batting
    """

    while True:
        if user_is_batter:
            run = int(input("Enter a number (0–6): "))
            bowl = random.randint(0, 6)
            print("Computer bowled:", bowl)
        else:
            bowl = int(input("Enter a number (0–6): "))
            run = random.randint(0, 6)
            print("Computer scored:", run)

        # OUT Condition
        if run == bowl:
            print("OUT!")
            break
        else:
            # Add the scoring run
            batter_array.append(run)


# TOSS DECISION

if toss == my_option:
    print("You win the toss and bat first!")

    # You bat first
    bat_first(my_run_array, computer_run_array, user_is_batter=True)

    print("\n--- Computer Now Batting ---")
    bat_first(computer_run_array, my_run_array, user_is_batter=False)

else:
    print("Computer wins the toss and bats first!")

    # Computer bats first
    bat_first(computer_run_array, my_run_array, user_is_batter=False)

    print("\n--- You Now Batting ---")
    bat_first(my_run_array, computer_run_array, user_is_batter=True)


# RESULT CALCULATION

my_total = sum(my_run_array)
computer_total = sum(computer_run_array)

print("\nYour Total:", my_total)
print("Computer Total:", computer_total)

if my_total > computer_total:
    print("You won by", my_total - computer_total, "runs!")
elif computer_total > my_total:
    print("Computer won by", computer_total - my_total, "runs!")
else:
    print("Match Tied!")
```
